```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import sys


sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Bibliotecas importadas com sucesso.")
```

```python

try:
    df_original = pd.read_csv('Country-data.csv')
except FileNotFoundError:
    print("Erro: Arquivo 'Country-data.csv' não encontrado.")
    print("Por favor, baixe o dataset do Kaggle e coloque-o no mesmo diretório.")
else:
    print(f"Dataset carregado com sucesso.")
    print(f"Número total de países (linhas) no dataset: {df_original.shape[0]}")
    
    print("\n--- Informações das Colunas e Tipos de Dados ---")
    df_original.info()
    
    print("\n--- 5 Primeiras Linhas do Dataset ---")
    display(df_original.head())
```

```python
print("--- Gerando Boxplots para Análise da Faixa Dinâmica ---")

features = df_original.columns.drop('country')
df_features = df_original[features]

n_features = len(features)
n_cols = 3
n_rows = int(np.ceil(n_features / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5))
axes = axes.flatten()

for i, col in enumerate(features):
    sns.boxplot(y=df_features[col], ax=axes[i], orient='v')
    axes[i].set_title(f'Distribuição de: {col}', fontsize=14)
    axes[i].set_ylabel('Valor')

for i in range(n_features, len(axes)):
    axes[i].set_visible(False)

plt.tight_layout()
plt.savefig('analise_faixa_dinamica.png')
plt.show()
```

### Análise Interpretativa da Faixa Dinâmica

**O que os resultados indicam?**

A análise dos boxplots (visível na saída acima) revela uma questão crítica: **as variáveis possuem faixas dinâmicas (escalas) drasticamente diferentes.**

* **Variáveis de Alta Magnitude:** `income` e `gdpp` (valores nas dezenas de milhares).
* **Variáveis de Média Magnitude:** `child_mort` (valores de ~0 a ~200).
* **Variáveis de Baixa Magnitude:** `inflation` e `total_fer` (muitos valores abaixo de 10).

**O que deve ser feito com os dados antes da etapa de clusterização?**

Algoritmos de clusterização baseados em distância (como K-Médias) são sensíveis a essas escalas. Sem tratamento, as variáveis `income` e `gdpp` iriam **dominar completamente** o cálculo da distância, fazendo com que o algoritmo ignorasse indicadores de saúde (ex: `child_mort`).

**Justificativa da Técnica:** É essencial aplicar a **Padronização (StandardScaler)** para transformar os dados (média 0, desvio padrão 1), garantindo que todas as variáveis contribuam de forma equilibrada para a clusterização.

```python
print("--- Iniciando Pré-processamento Adequado dos Dados ---")

print(f"Verificando valores ausentes por coluna:\n{df_original.isnull().sum()}")
print("\n(Dataset limpo, sem valores ausentes para tratar.)")

countries_labels_df = df_original[['country']]
df_features_to_scale = df_original.drop('country', axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_features_to_scale)

X_scaled_df = pd.DataFrame(X_scaled, columns=df_features_to_scale.columns)

print("\n--- Dados Pré-processados (5 primeiras linhas escalonadas) ---")
display(X_scaled_df.head())
```

```python
nome_arquivo_dados = 'dados_escalonados.csv'
nome_arquivo_labels = 'paises_labels.csv'

try:
    X_scaled_df.to_csv(nome_arquivo_dados, index=False)
    
    countries_labels_df.to_csv(nome_arquivo_labels, index=False)
    
    print(f"\n[SUCESSO] Pré-processamento concluído.")
    print(f"Arquivo de dados escalonados salvo como: '{nome_arquivo_dados}'")
    print(f"Arquivo de rótulos (países) salvo como: '{nome_arquivo_labels}'")

except Exception as e:
    print(f"\n[ERRO] Falha ao salvar arquivos CSV processados: {e}")
```
